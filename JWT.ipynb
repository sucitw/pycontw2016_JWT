```python
import json
import hmac
from hashlib import sha256
from base64 import urlsafe_b64encode
```

```python
segments = []
```

```python
header_dict = {
    'typ': 'JWT',
    'alg': 'HS256'
}
```

```python
json_header = json.dumps(header_dict).encode('utf-8')
```

```python
type(json_header)
```

```python
json_header
```

```python
header = urlsafe_b64encode(json_header)
```

```python
header
```

```python
segments.append(header)
segments
```

```python
payload_dict = {
    'user_id':'pythontw2016'
}
```

```python
json_payload =json.dumps(payload_dict).encode('utf-8')
```

```python
payload = urlsafe_b64encode(json_payload)
```

```python
payload
```

```python
segments.append(payload)
```

```python
segments
```

```python
SECRET =b'secret'
```

```python
signing_input =b'.'.join(segments)
signing_input
```

```python
sig = hmac.new (SECRET, signing_input, sha256)
```

```python
signature = urlsafe_b64encode(sig.digest())
```

```python
signature
```

```python
segments.append(signature)
segments
```

```python
token = b'.'.join(segments)
```

```python
token
```
